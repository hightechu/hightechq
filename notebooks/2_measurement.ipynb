```python
import numpy as np
from qiskit import *
%matplotlib inline
```

```python
# Create a Quantum Circuit acting on a quantum register of one qubit
cr = ClassicalRegister(1)
qr = QuantumRegister(1)
circ = QuantumCircuit(qr,cr)
# Add an X gate on qubit 0
circ.x(qr[0])

# Add a H gate, putting this in superposition
circ.h(qr[0])
circ.draw(output = 'mpl')
```

## Adding measurements

```python
circ.measure(qr, cr)
circ.draw(output='mpl')
```

## Let's make another circuit

```python
# Create a Quantum Circuit acting on a quantum register of theree qubits
cr = ClassicalRegister(3)
qr = QuantumRegister(3)
circ2 = QuantumCircuit(qr,cr)
circ2.x(qr[0])
circ2.h(qr[1])
circ2.x(qr[2])
circ2.h(qr[2])
circ2.draw(output = 'mpl')
```

## Add measurements

```python
circ2.measure(qr, cr)
circ2.draw(output='mpl')
```
